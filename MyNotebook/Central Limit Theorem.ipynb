# Central Limit Theorem
The central limit theorem (due to Laplace)tells us that, subject to certain mild conditions, the sum of a set of random variables,which is of course itself a random variable, has a distribution that becomes increasingly Gaussian as the number of terms in the sum increases.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

uni = np.random.uniform(0, 1, 100)
binom = np.random.binomial(10, 0.25, 1000)
```

```python
plt.hist(uni)
plt.show()
```

```python
plt.hist(binom)
plt.show()
```

```python
uni = np.random.uniform(0, 1, 1000) + np.random.uniform(0, 1, 1000)
plt.hist(uni)
plt.show()
```

```python
binom = np.random.binomial(10, 0.25, 1000) + np.random.binomial(10, 0.25, 1000)
plt.hist(binom)
plt.show()
```

```python
np.mean(binom)
```

```python
mu = np.mean(binom)
sigma = np.std(binom)
norm = np.random.normal(mu, sigma, 10000)
count, norm_hist, ignored = plt.hist(binom, normed=True)
plt.plot(norm_hist, 1/(sigma*np.sqrt(2*np.pi))*np.exp(-((norm_hist-mu)**2/(2*sigma**2))), linewidth=2, color='r')
plt.show()
```
